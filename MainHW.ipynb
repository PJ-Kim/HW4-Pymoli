## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#define bin as 0-10,10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
bins = [0, 9, 14, 19, 24, 29, 34, 39, 99]
```

```python
#group labels
group_names = ['0-10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
```

```python
#binning criteria to labels
unique_players['age_bins'] = pd.cut(unique_players['Age'], bins, labels=group_names)
```

```python
#df: calculations and formatting Age Dem values
pd.options.display.float_format = '{:,.2f}'.format
unique_bins = unique_players.groupby(['age_bins'], as_index=False).SN.count()
unique_bins['Percentage of Players'] = unique_bins['SN'] / unique_bins['SN'].sum()*100
unique_bins = pd.DataFrame({"age bins": unique_bins[('age_bins')], "Total Count": unique_bins[('SN')], "Percentage of Players":unique_bins[('Percentage of Players')]})
unique_bins.set_index('age bins', inplace=True)
unique_bins.index.name = "   "
unique_bins
```
